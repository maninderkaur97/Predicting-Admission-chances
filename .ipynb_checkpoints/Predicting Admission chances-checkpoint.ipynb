# Predicting admission chances 

The Goal of this project is to find what features affect the admissin_chance the most. I will be applying different models to predict admission_chance using different features and find how admission_chance changes with different features and models.

The dataset "Graduate Admissions" was found on kaggle.com.

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import scale, PolynomialFeatures
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier, export_graphviz
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import confusion_matrix, mean_squared_error, precision_recall_fscore_support, accuracy_score

%matplotlib inline
```

# Linear Regression

```python
############################################################################################################################
### Algorithm: Linear Regression                                                                                         ###
###                                                                                                                      ###
### Features:            GRE Score, TOEFL Score, University Rating, LOR, SOP, CGPA, Research                             ###
### Predicting:          admission_chance                                                                                ###
###                                                                                                                      ###
### Cross-validation OR Training/test data sets: For this model, training and test data sets are used                    ###
###                                                                                                                      ###
### Classification/regression: This is a regression model                                                                ###
###                                                                                                                      ###
### Supervised/Unsupervised: Supervised Learning                                                                         ###
###                                                                                                                      ###
### Regularization methods: No, regularization methods were not used for this model                                      ###
###                                                                                                                      ###
### Model Evaluation: Used MSE to evaluate the model.                                                                    ###
###                                                                                                                      ###
### Visualization: Scatter Plots are used to visualize the results                                                       ###
###                                                                                                                      ###
### Interesting Findings: I trained 3 different models using Linear Regression to predict "admission_chance". The MSE    ###
###                       did not quite change a lot. It MSE change whithin the range from 0.00448 to 0.0050.            ###
###                                                                                                                      ###
###                       Model 1:                                                                                       ###
###                               features: GRE Score, TOEFL Score, University Rating, LOR, SOP, CGPA, Research          ###
###                               MSE = 0.004480695237193526                                                             ###
###                                                                                                                      ###
###                                                                                                                      ###
###                       Model 2:                                                                                       ###
###                               features: GRE Score, TOEFL Score, CGPA                                                 ###
###                               MSE = 0.004667137970720909                                                             ###
###                                                                                                                      ###
###                                                                                                                      ###
###                       Model 3:                                                                                       ###
###                               features: SOP, LOR, CGPA, University Rating                                            ###
###                               MSE = 0.005010778052417498                                                             ###
###                                                                                                                      ### 
############################################################################################################################


X = data.drop(['admission_chance ', 'Serial No.'], axis = 1)
y = data['admission_chance ']


X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.5, random_state=1)


ln = LinearRegression()
ln.fit(X_train, y_train)

preds = ln.predict(X_test)
mse = mean_squared_error(y_test, preds)
print('MSE: ', mse)

plt.figure()
plt.scatter(y_test, preds)
plt.xlabel("Actual y Value")
plt.ylabel("Predicted y Value")
plt.show()
```

```python
X = data[['GRE Score', 'TOEFL Score', 'CGPA']]
y = data['admission_chance ']


X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.5, random_state=1)

lm2 = LinearRegression()
lm2.fit(X_train, y_train)

preds2 = lm2.predict(X_test)
mse = mean_squared_error(y_test, preds2)
print('MSE: ', mse)

plt.figure()
plt.scatter(y_test, preds)
plt.xlabel("Actual y Value")
plt.ylabel("Predicted y Value")
plt.show()
```

```python
X = data[['SOP', 'LOR ', 'CGPA', 'University Rating']]
y = data['admission_chance ']


X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.5, random_state=1)

lm3 = LinearRegression()
lm3.fit(X_train, y_train)

preds3 = lm3.predict(X_test)
mse = mean_squared_error(y_test, preds3)
print('MSE: ', mse)

plt.figure()
plt.scatter(y_test, preds)
plt.xlabel("Actual y Value")
plt.ylabel("Predicted y Value")
plt.show()
```

```python
############################################################################################################################
### Algorithm: Regression Tree                                                                                           ###
###                                                                                                                      ###
### Features:            GRE Score, TOEFL Score, University Rating, LOR, SOP, CGPA, Research                             ###
### Predicting:          admission_chance                                                                                ###
###                                                                                                                      ###
### Cross-validation OR Training/test data sets: For this model, training and test data sets are used                    ###
###                                                                                                                      ###
### Classification/regression: This is a regression model                                                                ###
###                                                                                                                      ###
### Supervised/Unsupervised: Supervised Learning                                                                         ###
###                                                                                                                      ###
### Regularization methods: No, regularization methods were not used for this model                                      ###
###                                                                                                                      ###
### Model Evaluation: Used MSE to evaluate the model.                                                                    ###
###                                                                                                                      ###
### Visualization: Scatter Plots are used to visualize the results                                                       ###
###                                                                                                                      ###
### Interesting Findings: I trained a models using Regression Tree to predict "admission_chance".                        ###
###                                                                                                                      ###
###                       Model 1:                                                                                       ###
###                               features: GRE Score, TOEFL Score, University Rating, LOR, SOP, CGPA, Research          ###
###                               MSE = 0.006927979772772137                                                             ###
###                                                                                                                      ###
###                                                                                                                      ###
###                                                                                                                      ### 
############################################################################################################################



X = data.drop(['admission_chance ', 'Serial No.'], axis = 1)
y = data['admission_chance ']


X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.5, random_state=1)

regr2 = DecisionTreeRegressor(max_depth=2)
regr2.fit(X_train, y_train)

pred = regr2.predict(X_test)

plt.scatter(pred, y_test, label='admission_chance ')
plt.plot([0, 1], [0, 1], '--k', transform=plt.gca().transAxes)
plt.xlabel('pred')
plt.ylabel('y_test')

MSE = mean_squared_error(y_test, pred)
print('MSE: ', MSE)
```

```python
############################################################################################################################
### Algorithm: Random Forest                                                                                             ###
###                                                                                                                      ###
### Features:            GRE Score, TOEFL Score, University Rating, LOR, SOP, CGPA, Research                             ###
### Predicting:          admission_chance                                                                                ###
###                                                                                                                      ###
### Cross-validation OR Training/test data sets: For this model, training and test data sets are used                    ###
###                                                                                                                      ###
### Classification/regression: This is a regression model                                                                ###
###                                                                                                                      ###
### Supervised/Unsupervised: Supervised Learning                                                                         ###
###                                                                                                                      ###
### Regularization methods: No, regularization methods were not used for this model                                      ###
###                                                                                                                      ###
### Model Evaluation: Used MSE to evaluate the model.                                                                    ###
###                                                                                                                      ###
### Visualization: Scatter Plots and  are used to visualize the results                                                  ###
###                                                                                                                      ###
### Interesting Findings: I trained a models using Regression Tree to predict "admission_chance".                        ###
###                                                                                                                      ###
###                       Model 1:                                                                                       ###
###                               features: GRE Score, TOEFL Score, University Rating, LOR, SOP, CGPA, Research          ###
###                               MSE = 0.005478948000000001                                                             ###
###                                                                                                                      ###
###                       Variable Importance:                                                                           ###
###                                I create a variable importance plot. It shows that GPA is the most important          ###
###                                feature. Better GPA will lead to higher admissin chance.                              ###
###                                                                                                                      ###
###                                                                                                                      ###
###                                In general, it was pretty                                                             ###
###                                obvious that you need better GPA for admission. I had also heard that your experience ###
###                                in research, strong letter of Purpose, and Letters of recommendation also play a big  ###
###                                role in getting admission for Masters program. According to the                       ###  
###                                Variable Importance Plot, this is not the case. They play a very minimum role.        ###
###                                                                                                                      ###
###                                                                                                                      ###
############################################################################################################################


X = data.drop(['admission_chance ', 'Serial No.'], axis = 1)
y = data['admission_chance ']


X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.5, random_state=1)

# Bagging: using all features
regr1 = RandomForestRegressor(max_features=4, random_state=1)
regr1.fit(X_train, y_train)

pred = regr1.predict(X_test)
plt.scatter(pred, y_test, label='admission_chance ')
plt.plot([0, 1], [0, 1], '--k', transform=plt.gca().transAxes)
plt.xlabel('pred')
plt.ylabel('y_test')
mean_squared_error(y_test, pred)
```

```python
Importance = pd.DataFrame({'Importance':regr1.feature_importances_*100}, index=X.columns)
Importance.sort_values(by='Importance', axis=0, ascending=True).plot(kind='barh', color='r', )
plt.xlabel('Variable Importance')
plt.gca().legend_ = None
```
